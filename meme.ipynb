```python
df = df.drop(["alt","facebookUrl","images/0/thumb","link","pageId","postId","time"],axis=1)
```

```python
df = df[df.isVideo != True]
```

```python
from datetime import datetime
df["timestamp"] = [datetime.fromtimestamp(x/1000).date() for x in df["timestamp"]]
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], format='%Y-%m-%d')
df = df.loc[(df["timestamp"] >= "2020-01-01")]
```
